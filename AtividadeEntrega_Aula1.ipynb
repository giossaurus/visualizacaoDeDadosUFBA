```python
import plotly.express as px
```

```python
df = px.data.iris()
df
```

```python
fig = px.bar(df, x='sepal_width', y='sepal_length')
fig.show()
```

```python
fig = px.bar(df, x='sepal_width', y='sepal_length', color='species')
fig.show()
```

```python
fig = px.bar(df, x='sepal_width', y='sepal_length', color='species', barmode='group')
fig.show()
```

```python
fig = px.scatter_3d(df,x='sepal_width', y = 'sepal_length', z = 'petal_length', color="species")
fig.show()
```

```python
df = px.data.tips()
df
```

```python
fig = px.pie(df, values='tip', names='day')
fig.show()
```
